# Sentiment Analysis

image.png

When we read a message from a friend, we can usually tell if they're happy, sad, or upset just from their words. But how can a computer understand these emotions? This is where AI comes in.

Think about how you learned to understand emotions in text. Maybe someone taught you that "I love this!" means the person is happy, while "This is terrible" means they're unhappy. AI models learn in a similar way, but they've been taught using millions of examples.

We're going to use something called sentiment analysis, which is just a fancy way of saying "emotion detection". The AI will look at a piece of text and tell us two important things:
1. Whether it thinks the text is positive or negative.
2. How confident it is about its decision.

This is particularly useful when you want to analyze lots of text quickly. Imagine reading through thousands of customer reviews to see if people are happy with a product—an AI can do this in seconds!

In the code that follows, we'll create an AI tool that can detect emotions in text. We'll test it with a simple happy sentence to see how it works. The AI will give us its best guess about the emotion in the text, along with a percentage showing how sure it is about its answer.

````python
# Create our mood detector
mood_detector = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment-latest")

# Try it with a simple happy sentence
text = "I love learning about AI!"
result = mood_detector(text)

print("\n=== AI Mood Analysis ===")
print(f"Text: {text}")
print(f"Mood: {result[0]['label']}")
print(f"Confidence: {result[0]['score']:.2%}")
````

When we run this code, the AI will tell us two things about our text:
- Whether it thinks the text is POSITIVE or NEGATIVE (this is the 'label').
- How confident it is about its decision (this is the 'score', shown as a percentage).

Let's try analyzing different kinds of sentences to see how our AI understands them:


````python
example_sentences = [
    "This is the best day ever!",
    "I don't like rainy days",
    "I'm so excited about learning Python!"
]

print("Let's see what mood the AI thinks these sentences have:\n")

for sentence in example_sentences:
    result = mood_detector(sentence)
    print("\n" + "="*50)
    print(f"📝 Text: {sentence}")
    print(f"😊/😢 Mood: {result[0]['label']}")
    print(f"📊 Confidence: {result[0]['score']:.2%}")
````

We'll use something called a text generation model. Think of it like giving AI the beginning of a story and asking it to write what happens next.

First, let's create our story generator:

````python
model_id = 'openai-community/gpt2'
#model_id = 'Qwen/Qwen2.5-3B'
story_maker = pipeline("text-generation", model=model_id)

# Let's try a simple story
prompt = "Once upon a time, there was a friendly dragon"
story = story_maker(prompt, max_length=50, num_return_sequences=1, truncation=True, pad_token_id=50256)
print("\n=== AI Story Time ===")
print(story[0]['generated_text'])
````

# Analyzing AI Story Generation Results

When working with AI story generation, you'll notice several common issues:

## Common Limitations to Watch For

### Coherence Problems
The AI might start with your prompt but then veer off in unexpected directions. This happens because the model generates text one word at a time without truly understanding the narrative structure. Watch for:
- Sudden topic changes.
- Contradictory statements.
- Lost plot threads.
- Characters that appear and disappear.

### Repetition and Patterns
AI models often fall into repetitive patterns. You might notice:
- Repeated phrases or ideas.
- Circular storytelling where the narrative doesn't progress.
- Overuse of common words or phrases.

### Lack of Causality
The model might generate events that don't logically follow from previous events. This shows the AI's limitation in understanding cause-and-effect relationships.

## Exercise: Critical Analysis
For each generated story, consider:
1. Does the story maintain logical consistency throughout?
2. Are character actions and events properly motivated?
3. Does the ending relate meaningfully to the beginning?
4. Would a human reader find the story satisfying?

Remember: These limitations don't mean the technology isn't useful—they mean we need to be thoughtful about how we use it. Consider AI-generated text as a starting point that requires human review and editing rather than a finished product.

# Teaching AI to Read and Answer Questions

Time to explore one of the most powerful abilities of AI—understanding text and answering questions about it! Just like you can read a story and answer questions about what happened, AI can do the same thing.

image.png

````python
# Let's create our AI question answerer
question_answerer = pipeline("question-answering")

# Here's an article for AI to read
article = """
Mars exploration has changed dramatically in the last 50 years. In 1976, Viking 1
became the first spacecraft to successfully land on Mars, sending back detailed
images of the red planet's surface. The most significant discovery came in 2013
when the Curiosity rover found evidence of ancient streambeds, proving that liquid
water once flowed on Mars. In 2021, the Perseverance rover landed in Jezero Crater,
an area scientists believe was once a river delta. This rover carries special tools
to search for signs of ancient microbial life and will collect rock samples that
will be returned to Earth in the future. The rover also brought along Ingenuity,
a small helicopter that made history by achieving the first powered flight on
another planet.
"""

# Let's ask a simple question first
simple_question = "Which rover landed in Jezero Crater?"
result = question_answerer(question=simple_question, context=article)

print("\n=== Simple Question Test ===")
print(f"Question: {simple_question}")
print(f"Answer: {result['answer']}")
print(f"Confidence: {result['score']:.2%}")
````

Notice how the AI found the exact piece of information it needed? Let's try some harder questions that require understanding more context:

````python
# Let's ask questions that need more understanding
complex_questions = [
    "What evidence suggests Mars once had water?",
    "What was the first successful achievement on Mars?",
    "What makes the Ingenuity helicopter historically significant?"
]

print("\n=== Complex Questions Test ===")
for question in complex_questions:
    answer = question_answerer(question=question, context=article)
    print(f"\nQ: {question}")
    print(f"A: {answer['answer']}")
    print(f"Confidence: {answer['score']:.2%}")
````

Let's understand what makes some questions harder than others. When we ask, "Which rover landed in Jezero Crater?" the answer is directly stated in the text. But when we ask about evidence for water on Mars, the AI needs to:

- Find the relevant information about streambeds.
- Understand that streambeds are evidence of water.
- Put this information together into an answer.

Let's try working with multiple pieces of information. Here's a new article about a different topic:

````python
nature_article = """
The Amazon rainforest acts as Earth's largest carbon sink. Every year, its trees
absorb billions of tons of CO2 from the atmosphere. However, recent studies show
that deforestation has reduced this capacity by 30% since 1990. Scientists warn
that if deforestation continues at the current rate, the Amazon might become
a carbon source instead of a sink by 2035.
"""

# Let's practice finding specific information
questions_about_amazon = [
    "What does the Amazon rainforest do with CO2?",
    "How has deforestation affected the Amazon?",
    "What might happen by 2035?"
]

print("\n=== Amazon Rainforest Analysis ===")
for question in questions_about_amazon:
    answer = question_answerer(question=question, context=nature_article)
    print(f"\nQ: {question}")
    print(f"A: {answer['answer']}")
    print(f"How sure is AI?: {answer['score']:.2%}")
````

# AI Translation: Breaking Language Barriers

We've learned how AI can understand and answer questions about text. Now, let's explore something even more powerful—teaching AI to translate between different languages! Translation AI helps people from different parts of the world understand each other.

image.png

````python
# Create our translator
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-fr")

# Let's try a simple translation to French
text = "Artificial intelligence is changing the world"
result = translator(text)
print("\n=== Simple Translation ===")
print(f"English: {text}")
print(f"French: {result[0]['translation_text']}")
````

One cool thing about AI translation is that it understands context. It's not just replacing words one by one—it tries to understand the meaning and express it naturally in another language. Let's see this in action:

````python
# Let's try some phrases that need context
phrases = [
    "I'm over the moon",  # idiom
    "The weather is cold but nice",  # contrasting ideas
    "Would you like to grab a coffee?",  # casual expression
]

print("\n=== Context-Aware Translation ===")
for phrase in phrases:
    result = translator(phrase)
    print(f"\nEnglish: {phrase}")
    print(f"French: {result[0]['translation_text']}")
````

We can even translate back and forth between languages! Let's create a function that translates text to French and back to English to see if the meaning stays the same:

````python
# Create translators for both directions
en_to_fr = pipeline("translation", model="Helsinki-NLP/opus-mt-en-fr")
fr_to_en = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")

def translate_roundtrip(text):
    # English to French
    french = en_to_fr(text)[0]['translation_text']

    # French back to English
    back_to_english = fr_to_en(french)[0]['translation_text']

    return {
        "original": text,
        "french": french,
        "back_to_english": back_to_english
    }

# Let's test it
test_text = "The early bird catches the worm"
result = translate_roundtrip(test_text)

print("\n=== Round Trip Translation ===")
print(f"Original: {result['original']}")
print(f"In French: {result['french']}")
print(f"Back to English: {result['back_to_english']}")
````

# AI That Can Summarize Text

Sometimes we have a long piece of text and want to quickly understand the main points. AI can help us create short summaries! This is super useful when you have long articles, stories, or documents to read.

# AI That Can Fill in the Blanks

Imagine having a friend who can guess missing words in sentences. That's what this AI can do! It's like playing a fill-in-the-blank game, but the AI uses its understanding of language to make smart guesses.

````python
# Create our text completion AI
fill_blank = pipeline("fill-mask", model="bert-base-uncased")

# Let's try a simple example
text = "The capital of France is [MASK]."
results = fill_blank(text)

print("\n=== Simple Completion ===")
print(f"Original text: {text}")
print("\nTop 3 guesses:")
for r in results[:3]:
    print(f"- {r['token_str']} ({r['score']:.2%} confident)")
````

The AI gives us multiple guesses, ordered by how confident it is about each one. Let's try something more interesting:

````python
# Try different types of sentences
sentences = [
    "I love eating [MASK] for breakfast.",
    "In summer, the weather is usually [MASK].",
    "Students use [MASK] to write their homework."
]

print("\n=== Multiple Examples ===")
for sentence in sentences:
    print(f"\nSentence: {sentence}")
    results = fill_blank(sentence)
    print("Top 3 guesses:")
    for r in results[:3]:
        print(f"- {r['token_str']} ({r['score']:.2%} confident)")
````
