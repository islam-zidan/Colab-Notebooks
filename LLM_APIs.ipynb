````python
import requests
import json
from google.colab import userdata

# Load your API key
groq_api_key = userdata.get('GROQ_API_KEY')

# Define the API endpoint
url = "https://api.groq.com/openai/v1/chat/completions"

# Set up headers
headers = {
    "Authorization": f"Bearer {groq_api_key}",
    "Content-Type": "application/json"
}

# Define the model and input prompt
data = {
    "model": "llama3-70b-8192",  # Groq's model name might slightly differ; check official docs
    "messages": [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "What is the capital of France?"}
    ],
    "temperature": 0.7
}

# Make the API request
response = requests.post(url, headers=headers, json=data)

# Parse and print the response
if response.status_code == 200:
    result = response.json()
    print(result["choices"][0]["message"]["content"])
else:
    print("Error:", response.status_code, response.text)
````

# Deep Dive: Understanding the API Code Structure

## The Building Blocks

The `requests` library is our primary tool for making HTTP requests to the API. It's like a messenger that carries our questions to the AI and brings back the answers. We use it to send POST requests, which is how we transmit data to the API.

The `json` library helps us work with JSON (JavaScript Object Notation) data. JSON is a standard way of formatting data that both humans and computers can easily understand. When we send requests to the API or receive responses, the data is formatted in JSON.


## Setting Up the Connection: API Configuration

The API configuration is like setting up the rules for how we'll communicate with the AI service. Let's break down each component:

```python
url = "https://api.groq.com/openai/v1/chat/completions"

headers = {
    "Authorization": f"Bearer {groq_api_key}",
    "Content-Type": "application/json"
}
```

The URL endpoint is like the AI's address—it's where we send our requests. In this case, we're using Groq's chat completions endpoint, which is designed for back-and-forth conversations with the AI.

The headers are like an envelope for our message. They contain two crucial pieces of information:
- The Authorization header proves our identity using our API key. The "Bearer" prefix is a standard way of indicating that we're using token authentication.
- The Content-Type header tells the API that we're sending JSON data. This is crucial because the API needs to know how to interpret our message.

## Crafting the Message: Request Structure

The request structure is where we format our message to the AI. This needs to be precise, as it determines how the AI will understand and respond to our query:

```python
data = {
    "model": "llama3-70b-8192",
    "messages": [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "What is the capital of France?"}
    ],
    "temperature": 0.7
}
```

Let's examine each component:

The `model` parameter specifies which AI model we want to use. Different models have different capabilities and specializations. The "llama3-70b-8192" model is one of Groq's powerful language models.

The `messages` list represents a conversation history. Each message has two parts:
- The `role` indicates who is speaking. There are three possible roles:
  - "system": These messages set up the AI's behavior and personality. Think of them as giving the AI its instructions.
  - "user": These messages contain our actual questions or requests.
  - "assistant": These appear in the response and contain the AI's replies.

The `temperature` parameter is fascinating—it controls how creative or focused the AI's responses will be:
- A temperature of 0.0 makes the AI very focused and deterministic, always giving similar responses to the same question.
- A temperature of 1.0 makes the AI more creative and varied in its responses.
- 0.7 is a sweet spot that balances creativity with consistency.

## Processing the Response: Response Handling

After we send our request, we need to handle the response properly:

```python
response = requests.post(url, headers=headers, json=data)

if response.status_code == 200:
    result = response.json()
    print(result["choices"][0]["message"]["content"])
else:
    print("Error:", response.status_code, response.text)
```

The response handling is like opening a package and checking its contents. Here's what's happening:

First, we check the status code. HTTP status codes are like receipt confirmations:
- 200 means "Success! Everything worked perfectly".
- 400 means "Bad Request" (something was wrong with our message).
- 401 means "Unauthorized" (problem with our API key).
- 429 means "Too Many Requests" (we're sending too many messages too quickly).
- 500 means "Server Error" (something went wrong on the API's end).

If we get a 200 status code, we:
1. Convert the response from JSON format using `response.json()`.
2. Navigate through the response structure to find the AI's message.
3. Extract and print the content of that message.

The response structure looks like this:
```python
{
    "choices": [
        {
            "message": {
                "role": "assistant",
                "content": "The capital of France is Paris."
            }
        }
    ]
}
```

If we get any other status code, we print both the code and the error message to help us understand what went wrong.

Production models are the most reliable and stable options in Groq's catalog. These models have been thoroughly tested and optimized for real-world applications. When working on important projects or assignments, you should use these models.

### Understanding Key Model Specifications

When selecting a model, pay attention to these important specifications:

1. Context Window
   - This represents how much text the model can process at once.
   - Measured in tokens (roughly 4 characters per token).
   - For example, a context window of 8,192 tokens can handle approximately 32,000 characters.
   - Larger context windows allow the model to maintain more conversation history.

2. Max Completion Tokens
   - This indicates the maximum length of the response the model can generate.
   - Important when you need longer, more detailed responses.
   - For example, llama-3.1-8b-instant has a max completion of 8,192 tokens.

3. Developer
   - Shows which organization created the base model.
   - Different developers optimize their models for different use cases.
   - Meta's models (like Llama) are known for general-purpose use.
   - Google's models often excel at technical tasks.
   - Mistral's models are optimized for efficient processing.

2. Context Needs

# Making API Calls with OpenRouter

Now that we understand how to access OpenRouter and its available models, let's examine how to make API calls to these models. The process shares many similarities with what we learned about the Groq API but includes some important OpenRouter-specific elements.

## Understanding the OpenRouter Code Structure

The code structure follows our previous API pattern with a few key differences. OpenRouter requires additional header information and uses a slightly different endpoint structure. Let's examine a complete example that generates a haiku using Google's Gemini model.


````python
from google.colab import userdata
openrouter_api = userdata.get('OPENROUTER')
````

````python
import requests
import json
from google.colab import userdata

# Load API key
openrouter_api_key = userdata.get('OPENROUTER')  # Make sure the variable name matches

# Define the OpenRouter API endpoint
url = "https://openrouter.ai/api/v1/chat/completions"

# Set up headers
headers = {
    "Authorization": f"Bearer {openrouter_api_key}",
    "Content-Type": "application/json",
    "HTTP-Referer": "https://colab.research.google.com/",
    "X-Title": "Colab Notebook Request"
}

# Define the model and input prompt
data = {
    "model": "google/gemini-2.0-pro-exp-02-05:free",
    "messages": [
        {"role": "system", "content": "You are a helpful AI assistant."},
        {"role": "user", "content": "Can you write me a haiku?"}
    ],
    "temperature": 0.7
}

# Make the API request
response = requests.post(url, headers=headers, json=data)

# Parse and print the response
if response.status_code == 200:
    result = response.json()
    print(result["choices"][0]["message"]["content"])
else:
    print("Error:", response.status_code, response.text)
````

## Key Differences from Other APIs

OpenRouter's implementation requires several specific elements:

1. Additional Headers
   - The "HTTP-Referer" header identifies where the request comes from.
   - The "X-Title" header provides a description of the request's purpose.

2. Model Selection
   - Models are specified using the provider/model-name format.
   - Free models include ":free" at the end of their identifier.

# Building a Streamlit Chat Application with OpenRouter

In this section, we'll create an interactive web application using Streamlit that interfaces with OpenRouter's API to access Google's Gemini model. This application will provide a user-friendly chat interface for interacting with the AI.

## Understanding the Application Structure

Our Streamlit application consists of several key components:

## The Complete Application Code

When you run the following code in your Colab notebook, it will create a file named `app.py` with our Streamlit application:

````python
from pyngrok import ngrok
from google.colab import userdata
import os

# Update the Secrets section in the left sidebar with the token you copied
ngrok.set_auth_token(userdata.get('NGROK_AUTH_TOKEN'))
````

````python
def run_streamlit():
    # Kill any running Streamlit instances
    !killall ngrok
    os.system("fuser -k 8501/tcp")

    # Run the Streamlit app
    os.system("streamlit run app.py &>/dev/null&")

    # Create a new ngrok tunnel
    public_url = ngrok.connect(8501)
    print(f"🌟 Your Streamlit app is running at: {public_url.public_url}")
    return public_url.public_url
````

````python
%%writefile app.py
import streamlit as st
import requests
import json
import os

# Initialize session state for chat history
if 'messages' not in st.session_state:
    st.session_state.messages = []

def send_message_to_openrouter(message):
    url = "https://openrouter.ai/api/v1/chat/completions"

    headers = {
        "Authorization": "Bearer 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",
        "Content-Type": "application/json",
        "HTTP-Referer": "https://colab.research.google.com/",
        "X-Title": "Streamlit Chat App"
    }

    data = {
        "model": "google/gemini-2.0-pro-exp-02-05:free",
        "messages": [{"role": "system", "content": "You are a helpful AI assistant."}] +
                   [{"role": m["role"], "content": m["content"]} for m in st.session_state.messages] +
                   [{"role": "user", "content": message}],
        "temperature": 0.7
    }

    try:
        response = requests.post(url, headers=headers, json=data)
        response.raise_for_status()
        return response.json()["choices"][0]["message"]["content"]
    except Exception as e:
        return f"Error: {str(e)}"

# Streamlit UI
st.title("AI Chat Assistant")
st.write("This chat application uses the OpenRouter API to access Google's Gemini model.")

# Chat interface
for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.write(message["content"])

# User input
if prompt := st.chat_input("What would you like to ask?"):
    # Display user message
    with st.chat_message("user"):
        st.write(prompt)

    # Add user message to state
    st.session_state.messages.append({"role": "user", "content": prompt})

    # Get AI response
    with st.chat_message("assistant"):
        with st.spinner("Thinking..."):
            response = send_message_to_openrouter(prompt)
            st.write(response)

    # Add assistant response to state
    st.session_state.messages.append({"role": "assistant", "content": response})

# Add a clear chat button
if st.button("Clear Chat"):
    st.session_state.messages = []
    st.experimental_rerun()
````
